```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import plotly.express as px
```

```python
def classify_data(data_list, classification_dict):
    classified_data = {}
    remaining_data = []

    # Initialize empty lists for each category
    for category in classification_dict.keys():
        classified_data[category] = []

    # Classify each word in the data list
    for word in data_list:
        classified = False
        for category, category_words in classification_dict.items():
            if word in category_words:
                classified_data[category].append(word)
                classified = True
                break
        if not classified:
            remaining_data.append(word)

    return classified_data, remaining_data
```

```python
def append_dicts(dict1, dict2):
    result_dict = dict1.copy()  # Create a copy of the first dictionary

    for key, value in dict2.items():
        if key in result_dict:
            # Key exists, append unique values to the existing key
            if isinstance(value, list):
                for item in value:
                    if item not in result_dict[key]:
                        result_dict[key].append(item)
            elif value not in result_dict[key]:
                result_dict[key].append(value)
        else:
            # Key doesn't exist, add new key-value pair
            if isinstance(value, list):
                result_dict[key] = list(set(value))
            else:
                result_dict[key] = [value]

    return result_dict
```

```python
# Calculate average frame for each sign within each category
average_frame = df_expanded.groupby(['category', 'sign'])['total_frame'].mean().reset_index()

# Sort the dataframe by total_frame within each category
average_frame_sorted = average_frame.sort_values(['category', 'total_frame'], ascending=[True, False])

# Create a dictionary to store the top 5 signs for each category
top_5_signs_dict = {}

# Iterate over each category
for category in average_frame_sorted['category'].unique():
    # Select the top 5 signs for the current category
    top_5_signs_category = average_frame_sorted[average_frame_sorted['category'] == category].head(5)
    
    # Store the top 5 signs in the dictionary
    top_5_signs_dict[category] = top_5_signs_category['sign'].tolist()

# Print the resulting dictionary
for category, signs in top_5_signs_dict.items():
    print(f"{category}: {signs}")
```
